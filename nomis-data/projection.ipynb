# Using PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
pca_result1= pca.fit_transform(df1)
pca_result2= pca.fit_transform(df2)
pca_result3= pca.fit_transform(df3)
pca_result4= pca.fit_transform(df4)
pca_result5= pca.fit_transform(df5)
```

**extract PCA result for each ethnic group**

```python
pca0_result1 = pca_result1[:,0]
pca0_result2 = pca_result2[:,0]
pca0_result3 = pca_result3[:,0]
pca0_result4 = pca_result4[:,0]
pca0_result5 = pca_result5[:,0]

pca1_result1 = pca_result1[:,1]
pca1_result2 = pca_result2[:,1]
pca1_result3 = pca_result3[:,1]
pca1_result4 = pca_result4[:,1]
pca1_result5 = pca_result5[:,1]
```

```python
pca0_result1
```

```python
pca1_result1
```

**change the result to be dataframe**

```python
data_pca = pd.DataFrame(data=[df['geography'], df['geography code'], pca0_result1, pca1_result1, pca0_result2, pca1_result2, 
                             pca0_result3, pca1_result3, pca0_result4, pca1_result4,
                             pca0_result5, pca1_result5]).T

data_pca.columns= ["geography", "geography code","White_pca0", "White_pca1" , "Asian/Asian British_pca0", "Asian/Asian British_pca1", 
                  "Black/African/Caribbean/Black British_pca0", "Black/African/Caribbean/Black British_pca1",
                  "Mixed/multiple ethnic group_pca0", "Mixed/multiple ethnic group_pca1", "Other ethnic group_pa0", 
                  "Other ethnic group_pa1"]

data_pca
```

```python
data_pca.to_csv('pca_result.csv')
```

```python
import matplotlib.pyplot as plt
plt.scatter(pca_result1[:,0], pca_result1[:,1], c = 'r');
plt.scatter(pca_result2[:,0], pca_result2[:,1], c = 'c');
plt.scatter(pca_result3[:,0], pca_result3[:,1], c = 'g');
plt.scatter(pca_result4[:,0], pca_result4[:,1], c = 'b');
plt.scatter(pca_result5[:,0], pca_result5[:,1], c = 'k');
```

# Using UMAP

```python
import umap
```

```python
umap = umap.UMAP()
```

```python
umap_result1= umap.fit_transform(df1)
umap_result2= umap.fit_transform(df2)
umap_result3= umap.fit_transform(df3)
umap_result4= umap.fit_transform(df4)
umap_result5= umap.fit_transform(df5)
```

**extract UMAP result for each ethnic group**

```python
umap0_result1 = umap_result1[:,0]
umap0_result2 = umap_result2[:,0]
umap0_result3 = umap_result3[:,0]
umap0_result4 = umap_result4[:,0]
umap0_result5 = umap_result5[:,0]

umap1_result1 = umap_result1[:,1]
umap1_result2 = umap_result2[:,1]
umap1_result3 = umap_result3[:,1]
umap1_result4 = umap_result4[:,1]
umap1_result5 = umap_result5[:,1]
```

```python
umap0_result1
```

**change the result to be dataframe**

```python
data_umap = pd.DataFrame(data=[df['geography'], df['geography code'], umap0_result1, umap1_result1, umap0_result2, umap1_result2, 
                             umap0_result3, umap1_result3, umap0_result4, umap1_result4,
                             umap0_result5, umap1_result5]).T

data_umap.columns= ["geography", "geography code","White_umap0", "White_umap1" , "Asian/Asian British_umap0", "Asian/Asian British_umap1", 
                  "Black/African/Caribbean/Black British_umap0", "Black/African/Caribbean/Black British_umap1",
                  "Mixed/multiple ethnic group_umap0", "Mixed/multiple ethnic group_umap1", "Other ethnic group_umap0", 
                  "Other ethnic group_umap1"]

data_umap
```

```python
data_umap.to_csv('umap_result.csv')
```
